```python
import networkx as nx
import numpy as np
from scipy.stats import ks_2samp
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
# Function to fit the densification exponent (a)
def fit_densification_power_law(nodes, edges):
    # Perform linear regression on log-log scale
    # filter out infinities and zeros
    nodes = np.array(nodes)
    edges = np.array(edges)
    log_nodes = np.log(np.where(nodes > 0, nodes, 1)).reshape(-1, 1)
    log_edges = np.log(np.where(edges > 0, edges, 1))
    
    model = LinearRegression()
    model.fit(log_nodes, log_edges)
    
    a = model.coef_[0]  # Densification exponent
    return a

# Function to calculate edge-to-node ratio based on DPL
def calculate_edge_to_node_ratio(nodes, edges, exponent):
    return edges / (nodes ** exponent)

# Main comparison function
def compare_densification(G1, G2):
    # Compute snapshots for both graphs
    nodes_G1, edges_G1 = G1.nodes(), G1.edges()
    nodes_G2, edges_G2 = G2.nodes(), G2.edges()
    # Fit densification exponents
    exponent_G1 = fit_densification_power_law(nodes_G1, edges_G1)
    exponent_G2 = fit_densification_power_law(nodes_G2, edges_G2)    
    # Calculate edge-to-node ratios
    ratio_G1 = calculate_edge_to_node_ratio(nodes_G1, edges_G1, exponent_G1)
    ratio_G2 = calculate_edge_to_node_ratio(nodes_G2, edges_G2, exponent_G2)
    
    # Perform KS test
    ks_statistic = ks_2samp(ratio_G1, ratio_G2).statistic
    
    print(f"KS Statistic: {ks_statistic}")

    return ks_statistic

# Example: Generate two synthetic graphs and compare
G1 = nx.barabasi_albert_graph(100, 2)  # Graph 1
G2 = nx.barabasi_albert_graph(100, 3)  # Graph 2

ks_statistic = compare_densification(G1, G2)
```
